```python
A = np.random.standard_normal(50)
B = np.random.normal(50, .1, 50)
print(A)
plt.plot(A)
plt.show()
```

```python
data = {'A': A, 'B': B}
# data = A.B
df = pd.DataFrame(data, columns=['A', 'B'])
corrMatrix = df.corr()
print(corrMatrix)
print(corrMatrix.loc['A', 'B'])
```

```python
data2 = np.corrcoef(A, B)
print data2
```

```python
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
X = np.random.standard_normal(100)
Y = np.random.standard_normal(100)
plt.scatter(X, Y, s=30, c='red')
plt.legend(["test"])
plt.xlabel("This is the x axis")
plt.ylabel("This is the y axis")
plt.title("This is X vs Y")
plt.savefig("python_figure_1.jpg")
plt.show()
```
