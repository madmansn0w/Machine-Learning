```python
# x = range(1, 11)
x = np.linspace(-math.pi, math.pi, 50)
print x
print x.size
print len(x)
y=x
f = np.outer(np.cos(y),1/(1+(x**2)))
print f.size
plt.contour(x,y,f)
plt.show();
```

```python
plt.contour(x, y, f, levels=[.1, .2, .3, .4, .5])
plt.show()
```

```python
# g = np.outer(np.cos(y),1/(1+(x**2))).t
fa = np.subtract(f, f.transpose())/2
print fa.size
plt.contour(x, y, fa)
plt.show();
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

ax.contour3D(X, Y, fa, 100, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
plt.show();
ax.view_init(60, 35)
fig
```
